```python
import os
import numpy as np
import glob
from PIL import Image

def getCoord(image, height, width, std):
    mid = width // 2
    coor = []
    final = []
    cnt = 0

    for i in range(height):
        if image[i, mid, 0] == 255 and image[i, mid, 1] == 255 and image[i, mid, 2] == 255:
            continue

        if coor:
            if abs(coor[-1] - i) > std:
                final.append((coor[0], coor[-1]))
                cnt += 1
                coor = []
            else:
                coor.append(i)
        else:
            coor.append(i)

        if i == height-1:
            start, end = coor[0], coor[-1]
            final.append((start, end))
            cnt += 1

    return cnt, final

def Croptoon(path, save_dir, std=150):
    hap = 0
    i = 0
    tmp = 0
    chgFlag = 0
    startFlag = 1

    for file_path in glob.glob(path + '/*'):
        directory, filename = os.path.split(file_path)

        #공백제거
        filename = filename.lstrip()

        # #이미 있으면 패스
        # if filename in os.listdir('/content/drive/MyDrive/'):
        #   continue

        #타이틀 회차 떼고 전처리
        title_string = filename.split('_')[0]

        #확장자 떼고 전처리
        except_extension = filename.split('.')[0]

        #타이틀별 폴더생성 #웹툰마다 폴더 생성
        if chgFlag == 1 or startFlag == 1 :
          paths_by_webtoon = save_dir + '/' + title_string
          try:
            os.mkdir(paths_by_webtoon);
            print('make paths_by_webtoon -->', paths_by_webtoon)
          except FileExistsError: print("file exists, Skip make dir" )
          chgFlag = 0
          startFlag = 0

        image = np.asarray(Image.open(file_path))
        cnt, final = getCoord(image, image.shape[0], image.shape[1], std)
        hap += cnt
        for (start, end) in final:
            cropped = image[start:end, :]
            if cropped.shape[0] < 250:
                hap -= 1
                continue
            cropped = Image.fromarray(cropped,'RGB')
            #저장
            cropped.save(paths_by_webtoon + '/' + except_extension + '_' + str(i) + '.jpg', 'JPEG')
            i += 1
            print(f'{filename} cropped => {cnt} images')

        #다음 웹툰으로 넘어가는 여부
        if(tmp != title_string):
          chgFlag = 1
        tmp = title_string

    print(f'Total {hap} images cropped')


if __name__ == '__main__':
    webtoon_path = '/content/drive/MyDrive/빅데이터캡스톤/프젝/웹툰 크롤링된 이미지'  # 웹툰 이미지 경로 설정
    saving_path = '/content/drive/MyDrive/빅데이터캡스톤/프젝/전처리된 이미지'  # 크롭된 이미지 저장 경로 설정

    if '전처리된 이미지' not in os.listdir('/content/drive/MyDrive/'):
        os.mkdir(saving_path)

    for dir in os.listdir(webtoon_path):
        if(saving_path):
          Croptoon(webtoon_path, saving_path, std=150)
        else:
          # Handle the case where the element does not exist
          print(f'Image with not found')
          break;

    for toon in os.listdir(webtoon_path):
        print(toon)
```
